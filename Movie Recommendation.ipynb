```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
    
def convert1(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
    
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert1)
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.dropna(subset=['overview'])
```

```python
import re

def cleanText(st):
    new = re.compile('[^A-Za-z]+')
    st_ = re.sub(string= st, repl=' ', pattern = new)
    return st_
```

```python
cleanText('A....B A.....B')
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stopwords_ = stopwords.words('english')

stopwords_
```

```python
preprocessed_overviews = []
# tqdm is for printing the status bar
for sentence in tqdm(movies['overview'].values):
    if type(sentence) == str:
        sent = cleanText(sentence)
        
        sent = ' '.join(e for e in sent.lower().split() if e not in stopwords_ and e != ' ')
        
        preprocessed_overviews.append(sent.lower().strip())
    else:
        preprocessed_overviews.append("")
preprocessed_overviews[2]
```

```python
preprocessed_titles = []
# tqdm is for printing the status bar
for sentence in tqdm(movies['title'].values):
    if type(sentence) == str:
        sent = cleanText(sentence)
        
        sent = ' '.join(e for e in sent.lower().split() if e not in stopwords_ and e != ' ')
        
        preprocessed_titles.append(sent.lower().strip())
    else:
        preprocessed_titles.append("")
```

```python
movies['overview'] = preprocessed_overviews
movies['new_titles'] = preprocessed_titles
movies.head()
```

```python
def toString(list_):
    st = ""
    for item in list_:
        st += item
        st += " "
        
    return st.strip()
toString(movies['genres'][0])
```

```python
movies['genres'] = movies['genres'].apply(toString)
movies['keywords'] = movies['keywords'].apply(toString)
movies['cast'] = movies['cast'].apply(toString)
movies['crew'] = movies['crew'].apply(toString)
```

```python
movies['Final'] = (movies['genres'].str.lower() + " " 
                   + movies['keywords'].str.lower() + " "
                   # + movies['new_titles'].str.lower() + " " 
                   + movies['overview'].str.lower() + " " 
                   + movies['cast'].str.lower() + " " 
                   + movies['crew'].str.lower()).str.strip()
```

```python
movies.head()
```

```python
from collections import Counter
my_counter = Counter()
for sent in movies['Final'].values:
    my_counter.update(sent.split())
```

```python
dict(my_counter)
```

```python
from collections import Counter
my_counter_title = Counter()
for sent in movies['new_titles'].values:
    my_counter_title.update(sent.split())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
final_vector = CountVectorizer(vocabulary = list(dict(my_counter).keys()), 
            lowercase = False, binary = True)
final_vector.fit(movies['Final'].values)

print(final_vector.get_feature_names_out())

final_vector_bow = final_vector.transform(movies['Final'].values)

final_vector_bow.toarray()
```

```python
final_vector_title = CountVectorizer(vocabulary = list(dict(my_counter_title).keys()), 
            lowercase = False, binary = True)
final_vector_title.fit(movies['new_titles'].values)

print(final_vector_title.get_feature_names_out())

final_vector_bow_title = final_vector_title.transform(movies['new_titles'].values)

final_vector_bow_title.toarray()
```

```python
final_vector_bow = final_vector_bow.toarray()
final_vector_bow_title = final_vector_bow_title.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(sent):
    sent = [sent]
    query = final_vector.transform(sent).toarray()
    query_ = final_vector_title.transform(sent).toarray()
    # query = np.expand_dims(query, axis = 0)

    dic = {i:cosine_similarity(query, np.expand_dims(vec, axis = 0))[0][0] for i, vec in enumerate(final_vector_bow) }
    dic_ = {i:cosine_similarity(query_, np.expand_dims(vec, axis = 0))[0][0] for i, vec in enumerate(final_vector_bow_title) }
    # dic = {**dic, **dic_}
    distances = sorted(dic.items(), reverse = True, key = lambda x:x[1])
    distances_ = sorted(dic_.items(), reverse = True, key = lambda x:x[1])

    movie_ind = []

    for item in distances_:
        if item[1] == 0.0:
            continue
        movie_ind.append(item[0])
        
    for item in distances:
        if item[0] not in movie_ind:
            movie_ind.append(item[0]) 

    for item in movie_ind[0:5]:
        print(movies['title'][item])
```
